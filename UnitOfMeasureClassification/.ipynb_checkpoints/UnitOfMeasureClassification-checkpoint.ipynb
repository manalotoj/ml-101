```python
def printValues(file, category, prefix, values):
    for value in values:
        # 1 pf, 1 farad, etc.
        for i in range(1, 9):
            file.write("{},{} {}\n".format(category, i, prefix, value))
            file.write("{},{}{}\n".format(category, i, prefix, value))
            file.write("{},{} to {} {}\n".format(category, i, i+5, prefix, value))
        
        # 20 f, 20 farads, etc.
        for i in range(0, 120, 20):
            file.write("{},{} {}\n".format(category, i, value))
            file.write("{},{}{}\n".format(category, i, value))
            file.write("{},{} to {} {}\n".format(category, i, i+20, value))

        # 21 f, 42 farads, etc.            
        for i in range(100, 300, 21):
            file.write("{},{} {} {}\n".format(category, i, prefix, value))
            file.write("{},{} {}{}\n".format(category, i, prefix, value))   
            file.write("{},{} {} to {} {}\n".format(category, i, i+50, prefix, value))
            
        # 0.1 f, 0.2 farads, etc.           
        for i in range(1, 100, 7):
            # measure, 0.1 farads (no trailing zeros)
            file.write("{},{} {} {}\n".format(category, '{0:.1f}'.format(i * 0.1), prefix, value))
        
        for i in range(1,100, 10):
            # measure, 0.1farads (no trailing zeros)
            file.write("{},{} {}{}\n".format(category, '{0:.1f}'.format(i * 0.1), prefix, value))
            
            # measure, 0.010 f
            file.write("{},{} {} {}\n".format(category, '{0:0.3f}'.format(i * 0.01), prefix, value))
            # measure, 0.010f
            file.write("{},{}{}\n".format(category, '{0:0.3f}'.format(i * 0.01), prefix, value))
            
            # measure, .01 f (no leading zero, no trailing zeros)
            x = '{0:0.3f}'.format(i * 0.01)
            file.write("{},{} {} {}\n".format(category, x[1:len(x)], prefix, value))
            file.write("{},{} {}{}\n".format(category, x[1:len(x)], prefix, value))
            
        #for i in range(1, 100, 1):
            #file.write("{},{} {}\n".format(category, '{0:.3f}'.format(i * 0.1), value))
            #file.write("{},{}{}\n".format(category, '{0:.3f}'.format(i * 0.1), value))            
        #for i in range(0, 10, 1):
            #file.write("{},{} {}\n".format(category,'{0:1.3f}'.format(i * 0.015), value))
            #file.write("{},{}{}\n".format(category, '{0:1.3f}'.format(i * 0.015), value))
            #file.write("{},{} to {} {}\n".format(category, '{0:1.3f}'.format(i * 0.015), '{0:1.3f}'.format(i+10), value))
        #for i in range(0, 10, 1):
            #file.write("{},{} {}\n".format(category,i * 0.015, value))
            #file.write("{},{}{}\n".format(category, i, value))
            #file.write("{},{} to {} {}\n".format(category, i * 0.015, i+10, value))           
            
def buildValues(prefixes, values):
    newValues = []
    for prefix in prefixes:
        for value in values:
            newValues.append(prefix + " " + value)
            newValues.append(prefix + value)
    return newValues

def buildCapacitance():
    categories = ["farads","milli farads","micro farads","pico farads"]
    baseValues = ["f","F","farads", "Farads","farad", "Farad"]
    microValues = buildValues(["micro","μ"], baseValues) 
    milliValues = buildValues(["milli","m","M"], baseValues) 
    picoValues = buildValues(["pico","p","P"], baseValues)

    catDic = {"farads":baseValues, "milli farads":milliValues, "micro farads":microValues, "pico farads":picoValues}
    for cat in categories:
        printValues(file, cat, "capacitance", catDic[cat])

def buildVoltage():
    categories = ["volts","milli volts","micro volts"]
    baseValues = ["v","V", "voltage", "Voltage","volts", "Volts","volt", "Volt"]
    microValues = buildValues(["micro","μ"], baseValues) 
    milliValues = buildValues(["milli","m"], baseValues)

    catDic = {"volts":baseValues, "milli volts":milliValues, "micro volts":microValues}
    for cat in categories:
        printValues(file, cat, "voltage", catDic[cat])
        
def buildResistance():
    categories = ["ohms","kilo ohms","mega ohms"]
    baseValues = ["ohms","Ohms","ohm", "Ohm","Ω"]
    megaValues = buildValues(["mega","Mega","m"], baseValues) 
    kiloValues = buildValues(["kilo","Kilo","K","k"], baseValues)

    catDic = {"ohms":baseValues, "kilo ohms":kiloValues, "mega ohms":megaValues}
    for cat in categories:
        printValues(file, cat, "resistance", catDic[cat])    

file = open("testdata.csv","w", encoding="UTF-8") 
file.write("Category,Value\n")

buildVoltage()
buildResistance()
buildCapacitance()

file.close()
```
